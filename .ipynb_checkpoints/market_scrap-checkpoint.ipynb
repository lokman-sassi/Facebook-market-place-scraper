```python
#Import browser to interact with chrome for example or firefox
from splinter import Browser

# BeautifulSoup for parsing HTML
from bs4 import BeautifulSoup as soup

#Regular expression library
import re

import pandas as pd
import matplotlib.pyplot as plt
import time
```

```python
#In my case, i'm using chrome browser, so consider changing it to the one you are using
browser = Browser('chrome')

base_url = "https://www.facebook.com/marketplace/constantine/propertyforsale?"
    
minPrice = 200
maxPrice = 6000

url = f"{base_url}minPrice={minPrice}&maxPrice={maxPrice}&latitude=36.3644&longitude=6.6086&radius=25"
```

```python
browser.visit(url)
```

```python
scroll_count = 4
scroll_delay = 2
for _ in range(scroll_count):
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")
    time.sleep(scroll_delay)
```

```python
#Parsing the HTML
html = browser.html

#Creating a beautifulSoup object from the scraped HTML
market_soup = soup(html, 'html.parser')

#Check if the HTML was scrapped correctly
market_soup
```

```python
browser.quit()
```

```python
#We came with a conclusion that classes names in Facebook are static, they have the same name in every publication in the market place

#class name of title in market place is the same for the others
titles_div = market_soup.find_all('span', class_="x1lliihq x6ikm8r x10wlt62 x1n2onr6")
titles_list = [title.text.strip() for title in titles_div]

#class name of the price in market place is the same for the others
prices_div = market_soup.find_all('span', class_="x193iq5w xeuugli x13faqbe x1vvkbs x1xmvt09 x1lliihq x1s928wv xhkezso x1gmr53x x1cpjm7i x1fgarty x1943h6x xudqn12 x676frb x1lkfr7t x1lbecb7 x1s688f xzsf02u")
prices_list = [price.text.strip() for price in prices_div]

#class name of the location in market place is the same for the others
locations_div = market_soup.find_all('span', class_="x1lliihq x6ikm8r x10wlt62 x1n2onr6 xlyipyv xuxw1ft x1j85h84")
locations_list = [location.text.strip() for location in locations_div]

#class name of title in description place is the same for the others
descriptions_div = market_soup.find_all('span', class_="x193iq5w xeuugli x13faqbe x1vvkbs x1xmvt09 x1lliihq x1s928wv xhkezso x1gmr53x x1cpjm7i x1fgarty x1943h6x xudqn12 x3x7a5m x6prxxf xvq8zen xo1l8bm xzsf02u")
descriptions_list = [description.text.strip() for description in descriptions_div]

#class name of the url in market place is the same for the others
urls_div = market_soup.find_all('a', class_="x1i10hfl xjbqb8w x1ejq31n xd10rxx x1sy0etr x17r0tee x972fbf xcfux6l x1qhh985 xm0m39n x9f619 x1ypdohk xt0psk2 xe8uvvx xdj266r x11i5rnm xat24cr x1mh8g0r xexx8yu x4uap5 x18d9i69 xkhd6sd x16tdsg8 x1hl2dhg xggy1nq x1a2a7pz x1heor9g x1lku1pv")
urls_list = [url.get('href') for url in urls_div]

```

```python
titles_list
```

```python
#enhanced_locations_list is a list that contains elements without strings that contains only numbers
enhanced_locations_list = [element for element in locations_list if not element.isdigit()]

#final_locations_list is a list that contains elements that don't start with DA characters
final_locations_list = [element for element in enhanced_locations_list if not element.startswith('DA')]


final_locations_list
```

```python
# Add all values to a list of dictionaries
houses_list = []

#changing format of data to document type like JSON "Key:Value" type
for i, item in enumerate(titles_list):
    house_dict = {}
    
    house_dict["Title"] = titles_list[i]
    house_dict["Location"] = final_locations_list[i]
    house_dict["Price"] = int(re.sub(r'[^\d.]', '', prices_list[i]))
    house_dict["Description"] = descriptions_list[i]
    house_dict["URL"] = urls_list[i]
    houses_list.append(house_dict)
    
print(houses_list)
```

```python
houses_df = pd.DataFrame(houses_list)

# Set the display option to ensure that all characters in a column are shown
pd.set_option('display.max_colwidth', None)

houses_df.tail()
```

```python
houses_df['URL'] = 'https://www.facebook.com/' + houses_df['URL']
```
